```python
import random
import matplotlib.pyplot as plt
xn , yn = 0.05, 0
ordonnées , abscisses = [] , []
for r in range(30000):
    p = random.uniform(0,1)
    if p < 0.02 :
        x=0.5
        y=0.27*yn
    elif 0.02<p<=0.17:
        x=-0.139*xn+0.263*yn+0.57
        y=0.246*xn+0.224*yn-0.036
    elif 0.17<p<=0.30:
        x=0.17*xn-0.215*yn+0.408
        y=0.222*xn+0.176*yn+0.0893
    else:
        x=0.781*xn-0.034*yn+0.1075
        y=-0.032*xn+0.739*yn+0.27
    xn=x
    yn=y
    ordonnées +=[y]
    abscisses +=[x]
plt.plot(abscisses, ordonnées, 'r ,' )
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
xn , yn = 0.5, 0.0
ordonnées , abscisses = [] ,[]
for r in range(30000):
    p = random.uniform(0,1)
    if p < 0.1 :
        x=0.05*xn
        y=0.60*yn
    elif 0.1<p<=0.2:
        x=0.05*xn
        y=0.50-yn+1
    elif 0.2<p<=0.4:
        x=0.46*xn-0.32*yn
        y=0.39*xn+0.38*yn+0.6
   
    elif 0.4<p<=0.6:
        x=0.47*xn-0.32*yn
        y=0.17*xn+0.42*yn+1.00
    elif  0.6<p<=0.8:
        x=0.34*xn+0.28*yn
        y=(-0.25)*xn+0.45*yn+1.00
    elif 0.8<p<1:
        x=0.42*xn+0.26*yn
        y=(-0.35)*xn+0.31*yn+0.7
    ordonnées +=[y]
    abscisses +=[x]
    xn=x
    yn=y
plt.plot(abscisses, ordonnées, 'c.' )
plt.show()
```
